```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url="https://www.rmxsba.com/18735_1/"
```

```python
url2="https://www.rmxsba.com/18735_2/"
```

```python
html=urlopen(url)
```

```python
html2=urlopen(url2)
```

```python
bsobj = BeautifulSoup(html, "lxml")
```

```python
novel_list=bsobj.find("div",{"id":"novel18735"}).find_all("dd")
```

```python
bsobj2=BeautifulSoup(html2, "lxml")
```

```python
novel_list2=bsobj2.find("div",{"id":"novel18735"}).find_all("dd")
```

```python
novel_dict={}

for i in range(len(novel_list)):
    novel_dict[novel_list[i].find("a").text]=novel_list[i].find("a").attrs["href"]

for j in range(len(novel_list2)):
    novel_dict[novel_list2[j].find("a").text]=novel_list2[j].find("a").attrs["href"]
    
novel_dict
```

```python
website="https://www.rmxsba.com/"
```

```python
novel_dict['诡秘之主笔趣阁第1章 绯红']
```

```python
i=0
for key,value in novel_dict.items():
    all_website=website+novel_dict[key]
    novel=urlopen(all_website)
    bsnovel = BeautifulSoup(novel, "lxml")
    content= bsnovel.find("div",{"class":"content"})\
    .text.replace("\n\xa0\xa0\xa0\xa0","\n")\
    .replace("\xa0\xa0\xa0\xa0","\n")\
    .replace("您可以在百度里搜索“诡秘之主笔趣阁 热门小说吧(www.rmxsba.com)”查找最新章节！","")\
    .replace("诡秘之主笔趣阁最新章节地址：https://www.rmxsba.com/shu/18735.html","")\
    .replace("诡秘之主笔趣阁全文阅读地址：https://www.rmxsba.com/18735/","")\
    .replace("诡秘之主笔趣阁txt下载地址：https://www.rmxsba.com/txtxz/18735.html","")\
    .replace("诡秘之主笔趣阁手机阅读：https://m.rmxsba.com/18735/","")\
    .replace("为了方便下次阅读，你可以点击下方的\"收藏\"记录本次（第1章 绯红）阅读记录，下次打开书架即可看到！","")\
    .replace("喜欢《诡秘之主笔趣阁》请向你的朋友（QQ、博客、微信等方式）推荐本书，谢谢您的支持！！(www.rmxsba.com)","")\
    .replace("\u2022","")\
    .replace("\xac","")\
    .replace("\xad","")
    print(i)
    i=i+1
    
    with open("诡秘之主.txt","a+") as novel:
        novel.write(str(key))
        novel.write("\n")
        novel.write(content)
        novel.write("\n")
```
