````r
simulations=10000#Number of samples you repeatedly take
y=c(1,2,3,4,5)#random variables' values
pr=c(0.35,0.05,0.15,0.05,0.4)#probability for each random variables' values


simulate<-function(simulations,y,pr,n){
    
    i=1#iterator
    sum_y=c()#store sum of random variables
    mean_y=c()#store mean of random variables
    while (i<=simulations){
        tem_sample=sample(y,size=n,replace=TRUE,prob=pr)#generage samples
        sum_y<-append(sum_y,sum(tem_sample))#append to sum of random variables
        mean_y<-append(mean_y,mean(tem_sample))#append to mean of random variables
        i=i+1
        }
    
    #theory
    E=sum(y*pr)#expectation
    Var=sum(y^2*pr)-E^2#variance
    
    theory_sum_y_mean=round(n*E,5)#theory value of mean of sum of random variables
    theory_sum_y_var=round(n*Var,5)#theory value of variance of sum of random variables
    
    theory_mean_y_mean=round(E,5)#theory value of mean of mean of random variables
    theory_mean_y_var=round(Var/n,5)#theory value of variance of mean of random variables
    
    
    #histogram represents the sum of random variables, sample size=n
    options(repr.plot.width = 13, repr.plot.height = 18)#canvas size
    sum_y_hist<-hist(sum_y,
                     main=paste("a histogram represents the sum of random variables\nsample size =",n),
                     xlab="sum of random variables",
                     freq=FALSE)
    
    #theory sum of random variables curve
    theory_x<-seq(n*min(y),n*max(y),length.out=10000)
    theory_sum_y<-dnorm(theory_x,mean=theory_sum_y_mean,sd=sqrt(theory_sum_y_var))
    lines(theory_x,theory_sum_y,col="red",lwd=5)
    
    sum_y_mean=round(mean(sum_y),5)
    sum_y_var=round(var(sum_y),5)
    #add legend
    legend("topleft",
           title=paste("theoretical distribution ～N(",theory_sum_y_mean,",",theory_sum_y_var,
                       ")\nsample distribution ～N(",sum_y_mean,",",sum_y_var,")"),
           legend=c("theoretical curve of sample distribution"),
           col=c("red"),lty=1, cex=0.65,bty="o",
           text.font=4, bg='lightblue')
    
    
    #histogram represents the mean of random variables, sample size=n
    mean_y_hist<-hist(mean_y,
                     main=paste("a histogram represents the mean of random variables\nsample size =",n),
                     xlab="mean of random variables",
                     freq=FALSE)
    
    #theory mean of random variables curve
    theory_x1<-seq(min(y),max(y),length.out=10000)
    theory_mean_y<-dnorm(theory_x1,mean=theory_mean_y_mean,sd=sqrt(theory_mean_y_var))
    lines(theory_x1,theory_mean_y,col="red",lwd=5)
    
    mean_y_mean=round(mean(mean_y),5)
    mean_y_var=round(var(mean_y),5)
    #add legend
    legend("topleft",
           title=paste("theoretical distribution ～N(",theory_mean_y_mean,",",theory_mean_y_var,
                       ")\nsample distribution ～N(",mean_y_mean,",",mean_y_var,")"),
           legend=c("theoretical curve of sample distribution"),
           col=c("red"),lty=1, cex=0.65,bty="o",
           text.font=8, bg='lightblue')
    
    
   
}
````

````r
# ANSWER BLOCK

# Least squared estimator function
#assume that dataset is clean
lsq <- function(dataset, predictor){
    # INSERT YOUR ANSWER IN THIS BLOCK
    meanx=mean(dataset[,predictor])#mean x
    meany=mean(dataset$consc_lev)#mean y
    meanxy=mean(dataset[,predictor]*dataset$consc_lev)#mean x*y
    meanx2=mean(dataset[,predictor]^2)#mean x^2
    SSxx=length(dataset)*(meanx2-meanx^2)#SSxx
    SSxy=length(dataset)*(meanxy-meanx*meany)#SSyy
    # Get the final estimators
    beta_1 <- SSxy/SSxx#B1
    beta_0 <- meany-beta_1*meanx#mean y and mean x must in line
    
    # Return the results:
    return(paste0('E[consc_lev]=', beta_0,'+', beta_1,'*',predictor))
}
print(lsq(train, 'input'))
print(lsq(train, 'v_pyr'))
````

Give an explanation for these results based on consideration of the predictors involved.

````r
# ANSWER BLOCK

# Some variables that you would want to initialize
minimum_error =314159265358979323846
features =""

# CODE HERE
all_predictors=names(train)[1:10]

for(i in 1:length(all_predictors)){
    for (j in i+1:length(all_predictors)){
        if(is.na(all_predictors[i])==FALSE && is.na(all_predictors[j])==FALSE){#select 2 without repeated from all
            formula=paste("consc_lev","~",paste(all_predictors[i],"+",all_predictors[j]))
            model2RV=lm(formula,data=train)
            rmse2RV=rmse(model2RV,train)
            if(rmse2RV<minimum_error){#set rmse to small one
                minimum_error=rmse2RV
                features=paste(all_predictors[i],",",all_predictors[j])
            }
        }    
    }  
}

print(paste('The best features are', features, '; and the MSE is', minimum_error))
````

after building the best lightweight linear regression models with ONLY TWO predictors, the rMSE just increase little, which means that the predictor deleted just has little influence to consc_lev and the rest predictors have the major influence to consc_lev.
